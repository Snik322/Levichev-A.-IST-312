```python
import matplotlib.pyplot as plt

calvin = [[207.5], [239.15], [273.15], [307.15], [327.15], [340.15], [373.15]]
fahrenheit = [[-86.6], [-29.2], [32.0], [93.2], [129.2], [152.6], [212.0]]
plt.figure(figsize=(15,8), dpi=50)
plt.scatter(fahrenheit, calvin, label='входные данные', color='green', marker='$f$');
plt.xlabel('calvin')
plt.ylabel('fahrenheit')
plt.legend()
plt.grid(True)
plt.show()
for c,f in zip(fahrenheit, calvin):
  print(f'кельвин {f}= фаренгейт{c}')
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(fahrenheit, calvin)
lr.predict([[256], [123]])
lr.coef_
lr.intercept_
calvin_test = [[-50], [10], [30], [20], [10], [70], [87]]
fahrenheit_test= lr.predict(calvin_test)
fahrenheit_test
for c,f in zip(fahrenheit_test, calvin_test):
  print(f'фаренгейт {c} кельвин{f}')
import numpy as np
x_range = np.arange(0, 100)
y_range = (x_range + 459.67) * 5/9
plt.figure(figsize=(15,8), dpi=280)
plt.plot(x_range, y_range, label='уравнение', linewidth='1')
plt.scatter(fahrenheit, calvin, label='входные данные', color='green')
plt.scatter(calvin_test, fahrenheit_test, label='предсказанное значение', color='blue')
plt.xlabel('фаренгейт')
plt.ylabel('кельвин')
plt.legend()
plt.grid(True)
plt.show()
```
